```python
from pyvox.parser import VoxParser
from pyvox.writer import VoxWriter
```

```python
voxels = VoxParser("../MagicaVoxel-0.99.6.4-win64/vox/wmap_101.vox")
# voxels = VoxParser("../MagicaVoxel-0.99.6.4-win64/vox/decode.vox")
# voxels = VoxParser("../MagicaVoxel-0.99.6.4-win64/vox/test_io.vox")
# voxels = VoxParser("../MagicaVoxel-0.99.6.4-win64/vox/out.vox")
```

```python
models = voxels.parse()
```

```python
for k in models.__dict__:
    print(k)
```

```python
print(len(models._palette))
models._palette
```

```python
print(len(models.materials))
```

```python
for mat in models.materials:
    if mat != None:
        print(mat)
```

```python
glass = models.materials[-3]
```

```python
from pyvox.models import Material
```

```python
old = models.materials[-1]
models.materials[-1] = Material(old.id, old.type, old.bid, old.btype, glass.content)
```

# write

```python
writer = VoxWriter("../MagicaVoxel-0.99.6.4-win64/vox/out.vox", models)
writer.write()
```
